## Полносвязная нейронная сеть

Реализовать нейронную сеть, состоящую из двух полносвязных слоев и решающую задачу классификации на наборе данных из лабораторной работы 1.

```python
import numpy as np
import matplotlib.pyplot as plt

from scripts.classifiers.neural_net import TwoLayerNet

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) 
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'


def rel_error(x, y):
    """ returns relative error """
    return np.max(np.abs(x - y) / (np.maximum(1e-8, np.abs(x) + np.abs(y))))
```

1. Добавьте реализации методов класса TwoLayerNet. Проверьте вашу реализацию на модельных данных (Код приведен ниже).  

```python
input_size = 4
hidden_size = 10
num_classes = 3
num_inputs = 5

def init_toy_model():
    np.random.seed(0)
    return TwoLayerNet(input_size, hidden_size, num_classes, std=1e-1)

def init_toy_data():
    np.random.seed(1)
    X = 10 * np.random.randn(num_inputs, input_size)
    y = np.array([0, 1, 2, 2, 1])
    return X, y

net = init_toy_model()
X, y = init_toy_data()
```

# Обучение нейронной сети на реальном наборе данных (CIFAR-10, MNIST)

Загрузите набор данных, соответствующий вашему варианту.

Обучите нейронную сеть на ваших данных. 

При сдаче лабораторной работы объясните значения всех параметров метода train.

```python
import numpy as np
from scripts.data_utils import load_CIFAR10
from sklearn.model_selection import train_test_split


# Загрузка данных CIFAR-10
cifar10_dir = 'scripts/datasets/cifar-10-batches-py'
X_train, y_train, X_test, y_test = load_CIFAR10(cifar10_dir)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)

# Предобработка данных
X_train = X_train.reshape(X_train.shape[0], -1) # Преобразуем данные в строки
X_val = X_val.reshape(X_val.shape[0], -1) # То же самое для валидационного набора
X_test = X_test.reshape(X_test.shape[0], -1) # И для тестового набора

print('Train data shape: ', X_train.shape)
print('Train labels shape: ', y_train.shape)
print('Validation data shape: ', X_val.shape)
print('Validation labels shape: ', y_val.shape)
print('Test data shape: ', X_test.shape)
print('Test labels shape: ', y_test.shape)

input_size = 32 * 32 * 3
hidden_size = 50
num_classes = 10
net = TwoLayerNet(input_size, hidden_size, num_classes)

stats = net.train(
    X_train, y_train, X_val, y_val, # тренировачные и валидационные данные
    num_iters=1000,  # количество итераций оптимизации
    batch_size=200,  # количество тренировочных примеров, используемых на каждом шаге
    learning_rate=1e-4,  # начальная скорость обучения
    learning_rate_decay=0.95,  # уменьшение скорости обучения после каждой эпохи
    reg=0.25,  # сила регуляризации
    verbose=True  # включение вывода информации о процессе обучения
)

val_acc = (net.predict(X_val) == y_val).mean()
print('Validation accuracy: ', val_acc)
```

```python
best_val_acc = -1
best_net = None

# Примеры гиперпараметров для тестирования
hidden_size_variants = [50, 75, 100]
learning_rate_variants = [1e-4, 5e-4, 1e-3]
reg_strength_variants = [0.25, 0.5, 0.75]
num_iters_variants = [1000, 1500, 2000]
dropout_variants = [0.0, 0.5]

for hidden_size in hidden_size_variants:
    for learning_rate in learning_rate_variants:
        for reg_strength in reg_strength_variants:
            for num_iters in num_iters_variants:
                for dropout in dropout_variants:
                    # Создание и обучение сети с текущим набором гиперпараметров
                    net = TwoLayerNet(input_size, hidden_size, num_classes, dropout=dropout)
                    stats = net.train(X_train, y_train, X_val, y_val,
                                      num_iters=num_iters, batch_size=200,
                                      learning_rate=learning_rate, learning_rate_decay=0.95,
                                      reg=reg_strength, verbose=False)
                    
                    # Вычисление точности на валидационном наборе
                    val_acc = (net.predict(X_val) == y_val).mean()
                    print(f'Validation accuracy: {val_acc}, hidden: {hidden_size}, lr: {learning_rate}, reg: {reg_strength}, iters: {num_iters}, dropout: {dropout}')
                    
                    # Обновление лучшей модели
                    if val_acc > best_val_acc:
                        best_val_acc = val_acc
                        best_net = net
                        
print(f'Best validation accuracy: {best_val_acc}')

# Вычисление и вывод точности лучшей модели на тестовом наборе
test_acc = (best_net.predict(X_test) == y_test).mean()
print(f'Test accuracy: {test_acc}')
```
